csv

```python
import csv
```

```python
path_csv = '../data_raw/dados_empresaB.csv'
```

```python
dados_csv = []
with open(path_csv, 'r') as file: 
    spamreader = csv.DictReader(file, delimiter=',')
    for row in spamreader: 
        dados_csv.append(row)
```

```python
dados_csv[1]
```

```python
nome_colunas_csv = list(dados_csv[0].keys())
nome_colunas_csv
```

```python
num_colunas_csv = len(dados_csv[0].keys())
num_colunas_csv
```

```python
key_mapping = {
    'Nome do Item':'Nome do Produto',
    'Classificação do Produto':'Categoria do Produto',
    'Valor em Reais (R$)': 'Preço do Produto (R$)',
    'Quantidade em Estoque':'Quantidade em Estoque', 
    'Nome da Loja':'Filial', 
    'Data da Venda':'Data da Venda'}
```

List compreension

```python
new_dados_csv = [{key_mapping.get(old_key): value for old_key, value in old_dict.items()} for old_dict in dados_csv]
```

```python
new_dados_csv
```

```python
new_dados_csv = []
for old_dict in dados_csv:
    dict_temp = {}
    for old_key, value in old_dict.items():
        dict_temp[key_mapping[old_key]] = value
    new_dados_csv.append(dict_temp)
```

```python
print(new_dados_csv)
```

```python
len(dados_json)
```

```python
len(new_dados_csv)
```

```python
combined_list = []
```

```python
combined_list.extend(dados_json)
combined_list.extend(new_dados_csv)
```

```python
len(combined_list)
```

```python
nome_colunas = list(combined_list[0].keys())
nome_colunas
```

```python
path_dados_combinados = '../data_processed/dadoscombinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=nome_colunas)
    writer.writeheader()

    for row in combined_list:
        writer.writerow(row)
```

```python
nome_colunas = list(combined_list[-1].keys())
```

```python
print(nome_colunas)
```

```python
dados_combinados_tabela = [nome_colunas]
for row in combined_list:
    linha = []
    for coluna in nome_colunas:
        linha.append(row.get(coluna, 'Indisponível'))
    dados_combinados_tabela.append(linha)
    
```

```python
path_dados_combinados = '../data_processed/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file: 
    writer = csv.writer(file)
    writer.writerows(dados_combinados_tabela)
```
